```python
# Load in necessary packages
import numpy as np
import matplotlib.pylab as plt
```

```python
# Convert csv of responses into an array containing every response to every questions
responses = np.loadtxt('FF_Awards_2025_responses_formatted.csv', delimiter=',',dtype=str)
```

For generating array of all responses to a question:

```python
# Seperate responses by question

Question = 35

# Q_options has the no. of options for each question (index-matched e.g Question 1 has Q_options[1] options)
Q_options = [0,5,6,4,5,2,4,3,7,2,5,3,5,6,3,10,4,4,5,6,4,6,5,5,3,6,6,5,5,4,6,4,7,7,7,8,4,6,5,6,8,7,5,5,5,5,7,5,6,6]

start = sum(Q_options[0:Question])
end = sum(Q_options[0:Question+1])

Q = [[0 for x in range(Q_options[Question])] for y in range(16)]

Q_answers = []
Q_delete = []

Q_c=[0 for x in range(16)]

for i in range(16):
    Q[i] = responses[i][start:end]
    Q_answers.insert(i,Q[i][0])

Q_names = list(set(Q_answers))
count = [0 for x in range(len(Q_names))]

for i in range(len(Q_names)):
    count[i] = Q_answers.count(Q_names[i])
    
max_count = max(count)

while max_count <= 8:
    plt.bar(Q_names, count, align='center')
    plt.xticks(rotation=30)
    plt.show()
    min_vote = min(count)
    for i in range(len(count)):
        if count[i] == min_vote:
            Q_delete.append(Q_names[i])
    for i in range(len(Q_delete)):
        for j in range(len(Q_answers)):
            if Q_answers[j] == Q_delete[i]:
                Q_c[j] +=1
                Q_answers[j] = Q[j][Q_c[j]]
    Q_names = list(set(Q_answers))
    count = [0 for i in range(len(Q_names))]
    for i in range(len(Q_names)):
        count[i] = Q_answers.count(Q_names[i])
    max_count = max(count)
    if max_count == 8 and len(Q_names) == 2:
        plt.bar(Q_names, count, align='center')
        plt.xticks(rotation=30)
        plt.show()
        break
else:
    plt.bar(Q_names, count, align='center')
    plt.xticks(rotation=30)
    plt.show()
```
